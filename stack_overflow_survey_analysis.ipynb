Завдання 7. Географічний аналіз компенсації Python-розробників

```python
plt.style.use('seaborn-v0_8')

plt.rcParams['font.size'] = 11
plt.rcParams['axes.titlesize'] = 13
plt.rcParams['axes.labelsize'] = 11

df['worked_with_python'] = df['LanguageHaveWorkedWith'].str.contains(
    'Python', 
    case=False, 
    na=False
)

df_python = df[df['worked_with_python']]

df_python = df_python.dropna(subset=['ConvertedCompYearly'])

py_progers_by_country = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

print(py_progers_by_country)


stats = df_python.groupby('Country')['ConvertedCompYearly'].agg(['median', 'count'])

stats = stats[stats['count'] >= 50]

top10 = stats.sort_values(by='median', ascending=False).head(10)

plt.figure(figsize=(12,6))

plt.ylim(0, max(top10['median']) * 1.1)

top10 = top10.copy()

top10.index = top10.index.str.replace(
    'United States of America', 'USA', regex=False
)

top10.index = top10.index.str.replace(
    'United Kingdom of Great Britain and Northern Ireland', 'UK', regex=False
)

plt.bar(top10.index, top10['median'], color='mediumseagreen')

plt.title('Медіанна зарплата Python-розробників за країнами (топ-10, ≥50 респондентів)')
plt.xlabel('Країна')
plt.ylabel('Зарплата (USD)')

plt.xticks(range(len(top10.index)), top10.index, rotation=45)

for i, v in enumerate(top10['median']):
    plt.text(i, v, f"{int(v):,}", ha='center', va='bottom')
  

plt.tight_layout()
plt.show()
```

Завдання 9. Аналіз популярності Python по віковим категоріям 

```python
Age_Python = df.groupby('Age')['worked_with_python'].mean().reset_index()

Age_Python['worked_with_python'] *= 100

Age_Python = Age_Python.rename(columns={
    'worked_with_python': 'Python_percent',
    'Age': 'Age (years old)'
})

Age_Python['Python_percent'] = Age_Python['Python_percent'].round(1)

Age_Python['Age (years old)'] = (
    Age_Python['Age (years old)']
    .str.replace(' years old', '', regex=False)
    .str.replace(' years or older', '+', regex=False)
    .str.replace('Prefer not to say', 'No answer', regex=False)
)

print(Age_Python)

plt.figure(figsize=(10,5))

bars = plt.bar(
    Age_Python['Age (years old)'],
    Age_Python['Python_percent'],
    color='cornflowerblue',
    width=0.6
)

plt.title('Відсоток Python-розробників за віковими категоріями')
plt.xlabel('Вік (роки)')
plt.ylabel('Відсоток (%)')

plt.xticks(rotation=45)

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height}%',
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()
```

Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників

```python
q75 = df['ConvertedCompYearly'].quantile(0.75)

high_paid = df[df['ConvertedCompYearly'] > q75]

remote_high_paid = high_paid[high_paid['RemoteWork'] == 'Remote']

industries = remote_high_paid['Industry'].value_counts().reset_index(name='count')

print(industries)

top10 = industries.head(10)

top10 = top10.sort_values(by='count')


plt.figure(figsize=(13,6))

plt.barh(top10['Industry'], top10['count'], color='slateblue')

plt.title('Топ-10 галузей серед високооплачуваних віддалених працівників')
plt.xlabel('Кількість працівників')
plt.ylabel('Галузь');

for i, v in enumerate(top10['count']):
    plt.text(v + 1, i, str(v), va='center')

plt.tight_layout()
plt.show()
```
